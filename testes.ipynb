```python
class Mamifero:
    def __init__(self, specie="Canino"):
        self.specie = specie

    def specie(self):
        print(self.specie)
```

```python
class Animal(Mamifero):
    def __init__(self, som, specie="Canino"):
        super().__init__(specie)
        self.speak = som

    def speak(self):
        print(f"Animal {self.speak}")
```

```python
cachorro = Animal("Auuuuuu")
gato = Animal("Miauuuu", "Felino")
```

```python
cachorro.speak
```

```python
cachorro.specie
```

```python
gato.speak
```
